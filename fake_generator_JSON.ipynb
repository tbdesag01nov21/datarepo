```python
# FUNCS
# For quick empty list assignment
def mklist(n):
    for i in range(n):
        yield []
        
# To easily add every data to the correct list
def item_to_lists(item, lists):
    for i in range(len(item)):
        lists[i].append(item[i])
        
    return lists

# To easily add every data to the correct list
def dict_to_lists(item, names, lists):
    for name, empty_list in zip(names, lists):
        if name in item.keys():
            empty_list.append(item.get(name))
            
        elif name not in item.keys():
            empty_list.append(0)
            
    return None
```
